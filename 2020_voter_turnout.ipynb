```python
import pandas as pd
```

```python
df = pd.read_csv('2020_turnout_rates.csv')

df.head()
```

```python
df = df.drop(['State Abv', 'Overseas Eligible', 'Vote for Highest Office (President)'], axis=1)
df.head()
```

```python
df = df.drop('Official/Unofficial', axis=1)
df
```

```python
df = df.drop('Source', axis=1)
df.head(15)
```

```python
df.columns
```

```python
df = df.rename({
    'Total Ballots Counted (Estimate)': 'TotalBallots',
    'VEP Turnout Rate': 'TurnoutRate',
    'Voting-Eligible Population (VEP)': 'EligibleVoters',
    'Voting-Age Population (VAP)': 'VotingAgePopulation',
    '% Non-citizen': 'NonCitizen',
    'Total Ineligible Felon': 'TotalIneligibleFelon'
}, axis=1)
df.head()
```

```python
df = df.replace(',','', regex=True)
df.head()
```

```python
df.dtypes
```

```python
df[['TotalBallots', 'EligibleVoters', 'VotingAgePopulation', 'Prison', 'Probation', 'Parole', 'TotalIneligibleFelon']] =\
    df[['TotalBallots', 'EligibleVoters', 'VotingAgePopulation', 'Prison', 'Probation', 'Parole', 'TotalIneligibleFelon']].astype(int)
```

```python
df.dtypes
```

```python
df = df.replace('%', '', regex=True)
df.head()
```

```python
df[['TurnoutRate', 'NonCitizen']] = df[['TurnoutRate', 'NonCitizen']].astype(float)
```

```python
df.dtypes
```

```python
import pyodbc

# Establish connection to SQL Server
conn = pyodbc.connect('Driver={SQL Server};'
                     'Server=DESKTOP-VF62RP9;'
                      'Database=Projects;'
                      'Trusted_Connection=yes;'
                     )
```

```python
# Create table in SQL Server
cursor = conn.cursor()

cursor.execute('CREATE TABLE VoterTurnout (State nvarchar(20), TotalBallots int, TurnoutRate decimal(3,1), EligibleVoters int,\
    VotingAgePopulation int, NonCitizen decimal(3,1), Prison int, Probation int, Parole int, TotalIneligibleFelon int)')
```

```python
# Insert DataFrame into created table
for row in df.itertuples():
    cursor.execute('''
        INSERT INTO Projects.dbo.VoterTurnout (State, TotalBallots, TurnoutRate, EligibleVoters, VotingAgePopulation, NonCitizen,\
        Prison, Probation, Parole, TotalIneligibleFelon)
        VALUES (?,?,?,?,?,?,?,?,?,?)
    ''',
    row.State,
    row.TotalBallots,
    row.TurnoutRate,
    row.EligibleVoters,
    row.VotingAgePopulation,
    row.NonCitizen,
    row.Prison,
    row.Probation,
    row.Parole,
    row.TotalIneligibleFelon)
    
conn.commit()
```
