```python
import pandas as pd

df = pd.read_csv('raw_data.csv', header=2)
df.head(10)
```

```python
# Get rid of unnecessary columns
df = df.drop(['Total', 'Footnotes'], axis=1)
df.head(10)
```

```python
# We don't want the first row with total US data
df = df.iloc[1:, :]
df.head(10)
```

```python
# Make NA values read '0' instead
df = df.fillna('0', axis=1)
df.head(10)
```

```python
df.dtypes
```

```python
# Get rid of unnecessary syntax
df = df.replace('<', '', regex=True)
df.head(10)
```

```python
# Convert objects into numeric data types for SQL compatability
df[['White', 'Black', 'Hispanic', 'Asian', 'American Indian/Alaska Native', 'Native Hawaiian/Other Pacific Islander', 'Multiple Races']] =\
df[['White', 'Black', 'Hispanic', 'Asian', 'American Indian/Alaska Native', 'Native Hawaiian/Other Pacific Islander', 'Multiple Races']].astype(float)
df.dtypes
```

```python
# Convert decimals into percent
df[['White', 'Black', 'Hispanic', 'Asian', 'American Indian/Alaska Native', 'Native Hawaiian/Other Pacific Islander', 'Multiple Races']] =\
df[['White', 'Black', 'Hispanic', 'Asian', 'American Indian/Alaska Native', 'Native Hawaiian/Other Pacific Islander', 'Multiple Races']].applymap(lambda x: x*100)
df.head()
```

```python
# Rename columns for clarity 
df = df.rename({
    'Location': 'State',
    'American Indian/Alaska Native': 'NativeAmerican',
    'Native Hawaiian/Other Pacific Islander': 'NativePacificIslander',
    'Multiple Races': 'MultipleRaces'
}, axis=1)
df.head()
```

```python
# Rows 52-64 are unnecessary; drop them.
df = df.iloc[0:51, :]
df.tail(10)
```

```python
import pyodbc

# Establish connection to SQL Server
conn = pyodbc.connect('Driver={SQL Server};'
                     'Server=DESKTOP-VF62RP9;'
                      'Database=Projects;'
                      'Trusted_Connection=yes;'
                     )
```

```python
# Create new table in SQL Server
cursor = conn.cursor()

cursor.execute('CREATE TABLE StateRacialDemographics (State nvarchar(20), White decimal(3,1), Black decimal(3,1), Hispanic decimal(3,1),\
    Asian decimal(3,1), NativeAmerican decimal(3,1), NativePacificIslander decimal(3,1), MultipleRaces decimal(3,1))')
```

```python
# Insert Dataframe into created table
for row in df.itertuples():
    cursor.execute('''
        INSERT INTO Projects.dbo.StateRacialDemographics (State, White, Black, Hispanic, Asian, NativeAmerican, NativePacificIslander,\
        MultipleRaces)
        VALUES (?,?,?,?,?,?,?,?)
    ''',
    row.State,
    row.White,
    row.Black,
    row.Hispanic,
    row.Asian,
    row.NativeAmerican,
    row.NativePacificIslander,
    row.MultipleRaces)
    
conn.commit()
```
